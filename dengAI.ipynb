Preprocesamiento

```python
# Eliminar 'week_start_date' solo si estÃ¡ presente
if "week_start_date" in features_train.columns:
    features_train.drop(columns=["week_start_date"], inplace=True)

if "week_start_date" in features_test.columns:
    features_test.drop(columns=["week_start_date"], inplace=True)
```

```python
features_train = pd.get_dummies(features_train, columns=["city"], drop_first=True)
features_test = pd.get_dummies(features_test, columns=["city"], drop_first=True)
```

```python
X = features_train.drop(columns=["year", "weekofyear"])
y = df_train["total_cases"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

Modelos

```python
models = {
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "Gradient Boosting": GradientBoostingRegressor()
}
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    mae = mean_absolute_error(y_val, y_pred)
    print(f"{name} MAE: {mae}")
```
